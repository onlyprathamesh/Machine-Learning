```python
def compute_net_amount(transactionlog):
    netamount = 0
    transactions = transactionlog.split()

    for i in range(0, len(transactions), 2):
        transaction_type = transactions[i]
        amount = int(transactions[i + 1])

        if transaction_type == 'D':
            netamount += amount
        elif transaction_type == 'W':
            netamount -= amount

    return netamount

transactionlog = input("Enter the transaction log: ")
net_amount = compute_net_amount(transactionlog)
print("Net Amount:", net_amount)
```

```python
def is_valid_password(password):
    return 6 <= len(password) <= 12 and any(c.islower() for c in password) and any(c.isupper() for c in password) and any(c.isdigit() for c in password) and any(c in '$#@' for c in password)

def get_valid_passwords(passwords):
    return [password for password in passwords if is_valid_password(password)]


input_passwords = input("Enter a sequence of comma-separated passwords: ")
valid_passwords = get_valid_passwords(input_passwords.split(','))

if valid_passwords:
    print("Valid passwords:", ', '.join(valid_passwords))
else:
    print("No valid passwords found.")
```
