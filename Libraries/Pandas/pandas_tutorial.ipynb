```python
dataFrame = pd.DataFrame(np.random.rand(344,5))
print('type of dataFrame:', type(dataFrame))
dataFrame
```

```python
# more knowledge of data frames

dataFrame.dtypes
dataFrame.columns
dataFrame.index
```

```python
# Convert the dataframe into numpy array
dataFrame.to_numpy()
```

```python
# Index sorting

dataFrame.sort_index(axis=1, ascending=False)
```

```python
# Creating a duplicate data frame

# This below thing only creates pointer to the location in the database. it does not create duplicate
# if changes made in the newDataFrame dataFrame will also get affected 
# newDataFrame = dataFrame

newDataFrame = dataFrame.copy()

# now this will make a duplicate of data frame and changes made in new will not reflect in old
newDataFrame[0][0] = 1
newDataFrame
dataFrame
```

### LOC method to change the df items

```python
# changing items by this method is not a good practice
newDataFrame[0][0] = 1

newDataFrame.loc[0,0] = 123
newDataFrame
```

```python
# Droping column or row
newDataFrame.drop([0,1], axis=0) # this line will create a copy and drop the row but original will not affect
# newDataFrame  # can check by uncommenting this line

# to permanently update original data frame
# newDataFrame.drop([0,1], axis=1, inplace=True)
newDataFrame
```

```python
# Returns particular row and cloumn combination

newDataFrame.loc[[1,5],[3,4]]

# Particular column and all rows

newDataFrame.loc[:,[3,4]]


# Particular row and all columns

newDataFrame.loc[[1,5],:]
```

```python
# Complex queries in LOC

newDataFrame.loc[(newDataFrame[1]<0.1) & (newDataFrame[3] >0.1)]
```

### iloc

```python
newDataFrame
```

```python
# by using iloc we can access items without their specific column name or row name

newDataFrame.iloc[0,3]
newDataFrame.iloc[[0,3],[1,2]]
```
