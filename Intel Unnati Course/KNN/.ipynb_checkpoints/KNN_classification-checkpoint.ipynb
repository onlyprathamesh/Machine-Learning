```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
df=pd.read_csv("/content/IRIS.csv")
```

```python
df.head()
```

```python
df['sepal_length'].loc[100]
```

```python
df.describe().T
```

```python
df.isnull()
df.fillna(df.mean())
```

```python
df.describe().T
```

```python
scaler=MinMaxScaler()
scaled=scaler.fit(df[['sepal_length','sepal_width','petal_length','petal_width']])
```

```python
scaled=scaler.transform(df[['sepal_length','sepal_width','petal_length','petal_width']])
scaled
```

```python
X=pd.DataFrame(data=scaled,
             columns=['sepal_length','sepal_width','petal_length','petal_width'])
```

```python
y=df['species']
```

```python
from sklearn.neighbors import KNeighborsClassifier
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=30)
model=KNeighborsClassifier(n_neighbors=3,metric="euclidean")
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
test_conf_matrix = confusion_matrix(y_train,y_test)
test_conf_matrix
```

```python
pd.DataFrame(test_conf_matrix,columns=model.classes_,index=model.classes_)
```
